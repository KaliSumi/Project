```python
# Install findspark
!pip install -q findspark

# Initialize findspark and SparkSession
import findspark
findspark.init()

from pyspark.sql import SparkSession

# Start Spark session
spark = SparkSession.builder \
    .appName("Loan Data Analysis") \
    .getOrCreate()

spark.version


```

```python

loan_data_df = spark.read.parquet("Loan Data - 3000x .parquet")

loan_data_df.show()
# Display schema
loan_data_df.printSchema()

# Basic statistics
loan_data_df.describe().show()

# Count records
print(f"Total number of records: {loan_data_df.count()}")

# Group by loan type and calculate average loan amount
from pyspark.sql.functions import avg
loan_data_df.groupBy("Loan_ID").agg(avg("LoanAmount").alias("Average_Loan")).show()

from pyspark.sql.functions import when

loan_data_df = loan_data_df.withColumn(
    "loanAmmont_Level",
    when(loan_data_df.LoanAmount > 500, "High").otherwise("Low")
)

loan_data_df.select("LoanAmount", "loanAmmont_Level").show(50)

loan_data_df = loan_data_df.fillna({"LoanAmountt": 0})

```
