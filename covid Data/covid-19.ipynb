```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import datetime

df=pd.read_csv("covid_daily_cases_sample.csv")
df['Date'] =pd.to_datetime(df['Date'])
grouped = df.groupby(['Country', 'Date'])['DailyCases'].sum().reset_index()
country = grouped.groupby('Country')['DailyCases'].sum().nlargest(10).index
print(df.head(10))



top_data=grouped[grouped['Country'].isin(country)]
plt.figure(figsize=(15, 8))
sns.barplot(data=top_data, x='Country', y='DailyCases', hue='Country')
plt.title('top 10 countries with highest daily cases')
plt.xticks(rotation=45)
plt.xlabel('country')
plt.ylabel('daily cases')
plt.legend(title='country')
plt.tight_layout()
plt.show()





```
